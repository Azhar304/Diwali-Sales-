```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fd=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
fd.shape
```

```python
fd.head(5)
```

```python
fd.drop(['Status', 'unnamed1'], axis=1, inplace =True)
```

```python
fd.tail(5)
```

```python
fd.info()
```

```python
# it will return true or false //pd.isnull(fd)\\
pd.isna(fd).sum()
```

```python
fd.dropna(inplace=True)
```

```python
fd.shape
```

```python
fd['Amount']= fd['Amount'].astype('int')
```

```python
fd.info()
```

```python
fd.rename(columns={'Marital_Status': 'Marriage'})
```

```python
ax=sns.countplot(x='Gender', data=fd)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=fd.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales)
```

```python
ax=sns.countplot(x='Age', data=fd, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=fd.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=sales_age, x='Age Group', y='Amount')
```

```python
sales_states=fd.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending =False).head(10)
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(data=sales_states, x='State', y='Orders')
```

```python
fd.info()
```

```python
spend=fd.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(data=spend, x='State' , y= 'Amount')
```

```python
ax=sns.countplot(data=fd, x='Marital_Status', hue='Gender')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(data=fd , x='Occupation')
sns.set(rc={'figure.figsize': (20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sector=fd.groupby(['Occupation'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False)
sns.barplot(data=sector, x='Occupation', y='Amount')
```
